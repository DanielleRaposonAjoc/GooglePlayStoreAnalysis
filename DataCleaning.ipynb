# Data Cleaning

```python
import pandas as pd
```

```python
apps_df = pd.read_csv("Google-Playstore.csv")
```

```python
apps_df.info()
```

```python
apps_df.head(15)
```

```python
#checking how many na values
apps_df.isna().sum()
```

```python
#remove privacy policy, currency, developer website, developer email, installs, minimum installs, minimum android
apps_df=apps_df.drop(["Privacy Policy", "Installs","Minimum Installs","Currency","Minimum Android","Scraped Time","Developer Email","Developer Id","Developer Website","Released","Last Updated"],axis=1)
```

```python
apps_df.isna().sum()
```

```python
apps_df.info()
```

```python
apps_df=apps_df.dropna(subset=["App Name"])
```

```python
apps_df.isna().sum()
```

```python
apps_df.head(15)
```

```python
apps_df.columns=["name","id","category","rating","ratingCount","installs","free","price","size","contentRating","hasAds","hasAppPurchase","isEditorsChoice"]
```

```python
apps_df.to_csv("apps_cleaned.csv")
```
