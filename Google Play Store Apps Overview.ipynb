```python
import pandas as pd
import seaborn as sns
import datetime as dt
sns.set()
```

# Dataset Overview

```python
apps_df = pd.read_csv("Google-Playstore.csv")
```

```python
apps_df.head()
```

```python
apps_df.isna().sum()
```

```python
#see how many unique values for category
print("Number of Unique Values: ",len(apps_df.Category.unique()))
apps_df.Category.value_counts(sort=False).sort_index(ascending=False).plot.barh(figsize=(10,10))
```

```python
#see how many unique values for size
print("Number of Unique Values: ",len(apps_df.Size.unique()))
apps_df.Size.value_counts(sort=False)[0:20].sort_index().plot.barh()
```

```python
#see how many unique values for installs
print("Number of Unique Values: ",len(apps_df.Installs.unique()))
apps_df.Installs.value_counts(sort=False).sort_index().plot.barh()
```

```python
#see how many unique values for type
print("Number of Unique Values: ",len(apps_df.Free.unique()))
print("Paid: ", len(apps_df[apps_df.Free==False]), "Free: ",len(apps_df[apps_df.Free==True]))
apps_df.Free.value_counts(sort=False).sort_index().plot.barh()
```

```python
#see how many unique values for price
print("Number of Unique Values: ",len(apps_df.Price.unique()))
apps_df.Price.plot.hist()
```

```python
#see how many unique values for currency
print("Number of Unique Values: ",len(apps_df.Currency.unique()))
apps_df.Currency.value_counts(sort=False)[0:20].sort_index().plot.barh(figsize=(5,10))
```

```python
#see how many unique values for ad supported
print("Number of Unique Values: ",len(apps_df.Free.unique()))
print("No ads: ", len(apps_df[apps_df["Ad Supported"]==False]), "No ads: ",len(apps_df[apps_df["Ad Supported"]==True]))
apps_df["Ad Supported"].value_counts(sort=False).sort_index().plot.barh()
```

```python
#see how many unique values for content rating
print("Number of Unique Values: ",len(apps_df["Content Rating"].unique()))
#print("No ads: ", len(apps_df[apps_df["Ad Supported"]==False]), "No ads: ",len(apps_df[apps_df["Ad Supported"]==True]))
display(apps_df["Content Rating"].value_counts(sort=False).sort_index())
apps_df["Content Rating"].value_counts(sort=False).sort_index().plot.barh()
```
